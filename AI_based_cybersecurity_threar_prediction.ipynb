Give a code to plot the Boxplot for outliers

```python
numerical_cols = df.select_dtypes(include=np.number).columns
num_cols_to_plot = numerical_cols[:10] # Select the first 10 numerical columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(num_cols_to_plot):
    plt.subplot(2, 5, i + 1)
    sns.boxplot(y=df[col])
    plt.title(col)

plt.tight_layout()
plt.show()
```

Give me a code to find the Correlation of features with Label

```python
# Encode the target variable
df_encoded = pd.get_dummies(df, columns=['Android_Adware'], drop_first=True)

# Calculate the correlation with the encoded label
correlation_with_label = df_encoded.corr(numeric_only=True)['Android_Adware_Android_Adware'].sort_values(ascending=False)
display(correlation_with_label)
```

Give me a code for Top features difference visualization

```python
# Get the top features based on absolute correlation with the label (excluding the label itself)
top_features = correlation_with_label.drop('Android_Adware_Android_Adware').abs().sort_values(ascending=False).head(10).index

# Visualize the difference in these top features between the two classes
plt.figure(figsize=(15, 10))
for i, col in enumerate(top_features):
    plt.subplot(2, 5, i + 1)
    sns.boxplot(data=df, x='Android_Adware', y=col)
    plt.title(col)

plt.tight_layout()
plt.show()
```
